```python
# 内联渲染画图
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')# 让图片漂亮一点
plt.rcParams['figure.figsize'] = (15, 5)
```

# 1.1 读取CSV文件

## 使用read_csv函数读取CSV文件。它默认文件是是用逗号分隔的。

## 我们将使用来自Montréal的自行车数据。原始链接请参考http://donnees.ville.montreal.qc.ca/dataset/velos-comptage

## 文件里面是一个List，内容是每天有多少人在Montreal的7个自行车道上。

```python
broken_df = pd.read_csv('data/bikes.csv')
#我发现会报错，其中有非utf-8的字符，用excel打开后删除第一行的几个乱码就可以了
```

```python
# 查看前三行
broken_df.head(3)
```

## 我们发现这个数据乱， read_csv有一些选项可以让我们处理这些问题，下面我们将：
### 1. 用a来切割列标（column）
### 2. 用'lating1'编码，默认的是‘utf-8’
###  3. 解析‘Date’列的数据
###  4. 告诉它我们需要把日放在月份的前面
###  5. 将行标（index）为‘Date’

```python
fixed_df = pd.read_csv('data/bikes.csv', 
                       sep=';',              #切割符 
                       encoding='latin1',
                       parse_dates=['Date'], #根据‘Date’解析日期
                       dayfirst=True,        
                       index_col='Date'
                      )
fixed_df[:3]
```

# 1.2 选择一列

## 当你使用read_csv函数的时候，你获得了一个由行和列组成的数据框架（DateFrame）。和Python中字典一样，你可以用同样的方式获得DataFrame的列。
## 下面是一个例子：

```python
fixed_df['Berri 1']
```

# 1.3 画一列

## 仅仅需要在后面加个.plot()即可！！
## 我们发现一月二月三月很少有人骑车，因为很冷啊！

```python
fixed_df['Berri 1'].plot()
```

## 我们可以很简单地把所有列都画在一起。我们发现所有的自行车道的表现基本是一致的--如果天气不适合骑自行车，那在哪都是一样的。

```python
fixed_df.plot(figsize=(15, 10))
```

# 1.4 将上面的所有合并在一起。

```python
df = pd.read_csv('data/bikes.csv', sep=';', encoding='latin1', parse_dates=['Date'], dayfirst=True, index_col='Date')
df['Berri 1'].plot()
```
